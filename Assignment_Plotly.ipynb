```python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/hhbenraheem/dads-lab/main/5001-LAB/Maven%20Market.csv")
df
```

```python
df['Total Profit'] = df['Total Profit'].map(lambda x: x.lstrip('$').rstrip('aAbBcC'))
```

```python
df['Total Profit'] = pd.to_numeric(df['Total Profit'])
```

```python
df2 = pd.DataFrame([df.groupby("product_brand")["Total Profit"].sum()])
df2=df2.T
df2.reset_index(inplace=True)
df2
```

```python
fig = px.bar(df, x="product_brand", y="Total Profit", title="Topline Performance")

fig.show()
```

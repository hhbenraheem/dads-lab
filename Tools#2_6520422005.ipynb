```python
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objs as go
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/hhbenraheem/dads-lab/main/5001-LAB/EPL-2021-2022-dataset.csv')
```

```python
df.head()
```

```python
HomeTeam_win = df[df['FTR'] == 'H']

df2 = HomeTeam_win.groupby('HomeTeam')['HomeTeam'].count().reset_index(name = "HomeWins")

df2.rename(columns={"HomeTeam": "Team"}, inplace = True)

df2
```

```python
AwayTeam_win = df[df['FTR'] == 'A']
df3 = AwayTeam_win.groupby('AwayTeam')['AwayTeam'].count().reset_index(name = "AwayWins")
df3.rename(columns={"AwayTeam": "Team"}, inplace = True)
df3
```

```python
Total_win = pd.merge(df2, df3, on="Team")

Total_win['TotalWins'] = Total_win['HomeWins']+Total_win['AwayWins']

Total_win["Wins/Games"] = Total_win['TotalWins'] / 38

Total_win
```

## 1st Question: How Often Does the Home Team Win?

```python
df['Result'] = np.where(df['FTR'] == 'H','Home Team Win',
                          np.where(df['FTR'] == 'A', 'Away Team Win', "Draw"))
general = df.groupby('Result')['Result'].count().reset_index(name = "count")
fig1 = px.pie(general, values ='count', names ='Result', title='Premier League - results from 2021 - 2022', color = 'Result',
             color_discrete_map={'Home Team Win':'royalblue',
                                 'Away Team Win':'orangered',
                                 'Draw':'lightgreen'})
fig1.update_traces(textposition='inside', textinfo='label+percent', textfont_size=9)
fig1.update_layout(showlegend=False)
fig1.show()
```

## 2nd Question: How many goals are in the season?

```python
df['Year']=pd.DatetimeIndex(df['Date']).year
df['TotalGoal'] = df['FTHG'] + df['FTAG']
total_goal = df.groupby('Year')['TotalGoal'].agg('sum').reset_index(name = 'Sum of Goals')
fig2 = px.bar(total_goal, x="Year", y="Sum of Goals",  title = "Total Goal", text_auto=True)
fig2.update_traces(textfont_size=12, textangle=0, textposition="outside", cliponaxis=False)
fig2.update_xaxes(type='category')
fig2.update(layout_yaxis_range = [0,800])
fig2.show()
```

## 3rd Question: How many fouls are in the season?

```python
df['TotalFouls'] = df['HF'] + df['AF']
total_fouls =df.groupby('Year')['TotalFouls'].agg('sum').reset_index(name = 'Sum of Fouls')
fig3 = px.bar(total_fouls, x = 'Year', y = 'Sum of Fouls', title = 'Total Fouls', text_auto='0.3s')
fig3.update_traces(textfont_size=12, textangle=0, textposition="outside", cliponaxis=False)
fig3.update_xaxes(type='category')
fig3.update(layout_yaxis_range = [0,6000])
fig3.show()
```

## 4th Question: What is the number goals per game in the season?

```python
fig4 = px.histogram(df, x=(df['FTHG']+df['FTAG']), title="Number of Goals per Match",text_auto=True)
fig4.update_layout(
    xaxis_title="Number of Goals",
    yaxis_title="Number of Matches",
)
fig4.show()
```

## 5th Question: Which team has the most wins ratio in the season?

```python
wins_on_games = Total_win.sort_values(by ='Wins/Games', ascending=True)
fig5 = px.bar(wins_on_games, x='Wins/Games', y='Team', text_auto=True, title='The Most Win Ratio Team')
fig5.update_layout(yaxis_tickformat = '.0%')
fig5.show()
```
